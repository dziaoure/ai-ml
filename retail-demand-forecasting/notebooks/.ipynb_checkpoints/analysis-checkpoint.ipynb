## 1. Data Loading and Processing

```python
import sys
from pathlib import Path

# Add the project root to the system path so that local module imports
# do not throw exceptions
PROJECT_ROOT = Path('..').resolve()

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import pandas as pd
import numpy as np

# Local imports
from src.data import load_walmart_data
```

```python
DATA_DIR = Path('..') / 'data'

stores, features, sales = load_walmart_data(DATA_DIR)

# Pre-merge integrity check
assert sales[['Store', 'Date']].duplicated().sum() == 0
assert features[['Store', 'Date', 'IsHoliday']].duplicated().sum() == 0
assert stores[['Store']].duplicated().sum() == 0

# Merge the sales ad features dataframes
df = sales.merge(
    features, 
    on=['Store', 'Date', 'IsHoliday'],
    how='left',
    validate='many_to_one'
)

df = df.merge(
    stores,
    on='Store',
    how='left',
    validate='many_to_one'
)

# Time-series safety
df = df.sort_values(['Store', 'Date']).reset_index(drop=True)
df.head()
```

## 3. Target Variable Distribution (`Weekly_Sales`)

Let's plot our target variable (`Weekly_Sales`) to detect any skew, identify outliers, and decide whether additional transformation is needed.

```python
# Plot the target variable distribution
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df['Weekly_Sales'], bins = 20)
plt.title('Distribution of Weekly Sales')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')
plt.show()
```

We can see that most weekly sales are clustered on the left (lower-to-moderate sales), and a small number of weeks have very high sales. THose extreme weeks create a long tail to the right. THerefore, the distribution is **rught-kewed** (the direction of the tail in to the right). This is common in retail. 

## 4. Time Series Trend (Global View)

Let's aggregate weekly sales across all stores

```python
weekly_sales_total = df.groupby('Date')['Weekly_Sales'].sum().reset_index()

plt.figure()
plt.plot(weekly_sales_total['Date'], weekly_sales_total['Weekly_Sales'])
plt.title('Total Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Weekly Sales')
plt.show
```

The time series plot clearly shows the seasonality of sales. Most weekly sales follow a regular pattern, with huge spikes during periods of high demand (usually holidays, Christmas, New Year, Valentine day, etc.).

## 5. Seasonality Analysis (Yearly Pattern)

Next, let's output the sales seasonality by week of year.

```python
df['week_of_year'] = df['Date'].dt.isocalendar().week 
seasonality = df.groupby('week_of_year')['Weekly_Sales'].mean()

plt.figure()
plt.plot(seasonality.index, seasonality.values)
plt.title('Average Weekly Sales by Week of Year')
plt.xlabel('Week of Year')
plt.ylabel('Average Weekly Sales')
plt.show()
```

As can be seen, there is a strong uplift in sales in the fourth quarter. This is due to the holidays (Halloween, Thanksgiving, Christmas, New Year). THis means that seaonality must be explicitly modeled.

## 6. Holiday Impact

Having noticed the Q4 sales uplift, let's now compare holiday versus non-holiday weeks.

```python
holiday_sales = df.groupby('IsHoliday')['Weekly_Sales'].mean()

holiday_sales.plot(kind = 'bar')
plt.title('Average Weekly Sales: Holiday vs Non-Holiday')
plt.xlabel('Is Holiday (1 = "Yes", 0 = "No")')
plt.ylabel('Average Weekly Sales')
plt.show()
```

We can clearly see that the average weekly sales are higher during holiday weeks.

## 7. Store Type

Next, let's plot the demand by store type.

```python
store_type_sales = df.groupby('Type')['Weekly_Sales'].mean().sort_values()

plt.figure(figsize=(10,6))
store_type_sales.plot(kind = 'bar')
plt.title('Average Weekly Sales by Store Type')
plt.xlabel('Store Type')
plt.ylabel('Average Weekly Sales')
plt.show()
```

As can be seen, larger stores (**type A**) have higher sales than smaller ones (**type C**)

## 8. Store Size vs Sales (Correlation Check)

Let's now see whether there's a correlation between store size and sales

```python
plt.figure()
plt.scatter(df['Size'], df['Weekly_Sales'], alpha = 0.5)
plt.title('Store Size vs Weekly Sales')
plt.xlabel('Store Size')
plt.ylabel('Weekly Sales')
plt.show()
```

As can be seen, there is a positive correlation between the store size and the weekly sales. Larger stores generate higher weekly sales whereas smaller ones generate smaller sales.

## 9. Correlation Matrix (Numeric Features)

Having concluded that there is a positve correlation between the store size and weekly sales, let's now generate the correlation matrix for numeric features.

```python
numeric_columns = [
    'Weekly_Sales',
    'Temperature',
    'Fuel_Price',
    'CPI',
    'Unemployment',
    'Size'
]

corr = df[numeric_columns].corr()

plt.figure()
plt.imshow(corr)
plt.xticks(range(len(corr)), corr.columns, rotation=45)
plt.yticks(range(len(corr)), corr.columns)
plt.colorbar()
plt.title('Correlation Matrix')
plt.show()
```

The correlation analysis shows that macroeconomic variables such as **CPI**, **unemployment**, **fuel price**, and **temperature** exhibit relatively weak linear relationships with weekly sales. In contrast, **store size** displays a clear positive correlation with demand, reflecting its role as a structural capacity constraint rather than a macroeconomic driver.
